```python
import pandas as pd
import warnings
warnings.filterwarnings(action="default")
warnings.filterwarnings(action="ignore")
```

Explorando os dados

```python
df = pd.read_csv("data/train.csv")
df.head()
```

| **Coluna**      | **Descrição Resumida**                                                                 |
|-----------------|----------------------------------------------------------------------------------------|
| PassengerId     | ID único para cada passageiro.                                                          |
| Pclass          | Classe do passageiro (1 = primeira, 2 = segunda, 3 = terceira).                        |
| Name            | Nome do passageiro.                                                                     |
| Sex             | Gênero do passageiro (male = homem, female = mulher).                                  |
| Age             | Idade do passageiro (se disponível).                                                    |
| SibSp           | Número de irmãos/cônjuges a bordo com o passageiro.                                     |
| Parch           | Número de pais/filhos a bordo com o passageiro.                                         |
| Ticket          | Número do bilhete de embarque.                                                          |
| Fare            | Tarifa paga pelo passageiro pelo bilhete.                                               |
| Cabin           | Número da cabine onde o passageiro estava (se disponível).                             |
| Embarked        | Porto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton).                     |
| Survived        | Variável alvo (0 = não sobreviveu, 1 = sobreviveu).                                     |

Somatoria de valores nulos

```python
df.isnull().sum()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df['IsAlone'] = (df['FamilySize'] == 0).astype(int)
```

```python
df
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
# TODO: Ver se a mediana da idade com base nas relacoes familiares faz mais sentido.

df['Cabin'].fillna('Unknown', inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# TODO: Ver se faz sentido pegar a moda refetente ao tipo de classe da passagem


df['Fare'].fillna(df.groupby('Pclass')['Fare'].transform('median'), inplace=True)
```

Convertendo colunas categóricas 

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```
